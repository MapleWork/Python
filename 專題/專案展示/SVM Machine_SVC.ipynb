```python
df = pd.read_csv('Result_16N.csv', index_col=[0, 1], encoding='utf_8_sig')

X = df.drop(["現金股利", "ROE_5YF", "EPS_5YF", "毛利率_5YF", "現金殖利率_F", "股價_F", "董監持股_F", "Sum", "判斷"], axis=1)

y = df['判斷']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

model.fit(X_train, y_train)

y_pred=model.predict(X_test)

print('Model accuracy score with rbf kernel and C=1000.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
Data = pd.DataFrame(y_pred)
Data.columns = ["答案"]
Data.to_csv("Answer.csv", index=False, encoding='utf_8_sig')
Data
```
