```python
import os
import glob
import pandas as pd

def get_stocknumber():
    res = []
    try:
        with open('D:\\專題\\程式\\stock_number.csv', encoding="utf-8") as f:
            slist = f.readlines()
            print('讀入：',slist)
            for lst in slist:
                s = lst.split(',')
                res.append([s[0].strip(),str(s[1].strip()),str(s[2].strip())])
    except:
        print('讀不到')
    return res

def merge(path, html_path, number):

    for i in range(0,len(os.listdir(html_path))):
        folder = (os.listdir(html_path)[i])

        Path_CSV = path+"\\calculate\\"+folder+"\\"

        files = glob.glob(os.path.join(Path_CSV, "Stock_"+number+".csv"))

        for i in files:
            fr = open(i,'rb').read()
            with open(path+"\\calculate\\All.csv",'ab') as f:
                f.write(fr)
                
path = 'D:\\資料庫\\台股\\上市'
html_path = 'D:\\資料庫\\台股\\html\\上市'
                
slist = get_stocknumber()
cnt = len(slist)
for i in range(cnt):
    number,name,group = slist[i]
    try:
        merge(path, html_path, number)
        print(number + "\t Success")
    except:
        pass
```
