```python
import requests
import pandas as pd
```

```python
def get_income_statement(sym):
    
    url = 'https://mops.twse.com.tw/mops/web/ajax_t164sb04'
    payload = {
            'encodeURIComponent':'1',
            'step':'1',
            'firstin':'1',
            'off':'1',
            'queryName':'co_id',
            'inpuType':'co_id',
            'TYPEK':'all',
            'isnew':'true',
            'co_id': str(sym),
    }
    res = requests.post(url, data = payload)
    df = pd.read_html(res.text)[1]
    
    return df
```

```python
get_income_statement(2303)
```
