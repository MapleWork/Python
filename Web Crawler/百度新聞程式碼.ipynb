```python
import requests # 匯入requests函式庫以便能夠在後續程式碼中執行http請求
from bs4 import BeautifulSoup #從bs4函式庫匯入BeautifulSoup函式，用來解析網頁。


#---索取網頁內容---
response = requests.get("http://news.baidu.com/")
#以request函式庫內的get()函式來請求http網頁內容，其中括弧內通常傳入字串形態網址。
#完成上述網頁內容請求之後，將結果放入等號左邊的response變數存放。



#---網頁內容解析---
soup = BeautifulSoup(response.text, 'html.parser')
#透過BeautifulSoup()函式來解析請求回來的網頁內容，
#其中括弧內第一個參數response.text是解析對像、第二個參數html.parser是解析模式。
#最後將解析結果放到等號左邊的soup變數存



#---網頁內容定位---
news = soup.select('.mod-tab-content a') 
#藉由select()函式，從解析完成的soup中選取所欲爬取之標的物。
#並且將選取完成之標的放入等號左邊的news變數存放

clean_news = news[0:37] 
#由於news list中末兩筆資料並非所欲選取標的，因此僅納入0:37筆資料。
#最後將整理結果放入等號左邊的clean_news存放



#---網頁內容爬取---
for item in clean_news: #透過for迴圈來遍歷clean_news內的每一筆資料
    print(item.text) #列印出該筆資料的文字內容
    print(item['href']) #列印出該筆資料的超連結內容
```

```python
import requests # 匯入requests函式庫以便能夠在後續程式碼中執行http請求
from bs4 import BeautifulSoup #從bs4函式庫匯入BeautifulSoup函式，用來解析網頁。

response = requests.get("http://news.baidu.com/")
#以request函式庫內的get()函式來請求http網頁內容，其中括弧內通常傳入字串形態網址。
#完成上述網頁內容請求之後，將結果放入等號左邊的response變數存放。

soup = BeautifulSoup(response.text, 'html.parser')
#透過BeautifulSoup()函式來解析請求回來的網頁內容，
#其中括弧內第一個參數response.text是解析對像、第二個參數html.parser是解析模式。
#最後將解析結果放到等號左邊的soup變數存

news = soup.select('#pane-news a')
#另外一種作法，選擇更內層爬取標的來排除不要的資料 (如id)。

for item in news: #透過for迴圈來遍歷clean_news內的每一筆資料
    print(item.text) #列印出該筆資料的文字內容
    print(item['href']) #列印出該筆資料的超連結內容
```
